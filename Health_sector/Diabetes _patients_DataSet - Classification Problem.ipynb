```python
dataset['Glucose'] = np.where(dataset['Glucose']==0,117,dataset['Glucose'])
dataset['BloodPressure'] = np.where(dataset['BloodPressure']==0,72,dataset['BloodPressure'])
dataset['SkinThickness'] = np.where(dataset['SkinThickness']==0,23,dataset['SkinThickness'])
dataset['Insulin'] = np.where(dataset['Insulin']==0,30.5,dataset['Insulin'])
dataset['BMI'] = np.where(dataset['BMI']==0,32,dataset['BMI'])
```

```python
dataset.head()
```

# Imblance treatement
# split the data into ind and dep variable

```python
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
# imbalance treatement
y.value_counts()
```

```python
print(x.shape, y.shape)
```

# Oversampling method to Balance the dataset

```python
#!pip install imblearn
import imblearn
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler()
x_ros, y_ros = ros.fit_resample(x,y)
```

```python
print("imbalance data :", y.value_counts())
print()
print("balance data :", y_ros.value_counts())
```

```python
# new dataset is -
print(x_ros.shape, y_ros.shape)
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_ros, y_ros, test_size=0.25, random_state=101)
```

# RandomForest Classification and Bagging Classification model

```python
from sklearn.ensemble import BaggingClassifier
bagging = BaggingClassifier()
bagging.fit(x_train, y_train)
```

```python
y_pred_bg_train = bagging.predict(x_train)
y_pred_bg_test = bagging.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Training Accuracy :", accuracy_score(y_train, y_pred_bg_train))
print()
print("Test Accuracy :", accuracy_score(y_test, y_pred_bg_test))
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1)
rf.fit(x_train, y_train)
```

```python
y_pred_rf_train = rf.predict(x_train)
y_pred_rf_test = rf.predict(x_test)
```

```python
print("Training Accuracy :", accuracy_score(y_train, y_pred_rf_train))
print()
print("Test Accuracy :", accuracy_score(y_test, y_pred_rf_test))
```

```python
from sklearn.model_selection import cross_val_score
accuracy_train = cross_val_score(rf, x_train, y_train, cv=10)
accuracy_test = cross_val_score(rf, x_test, y_test, cv=10)
print(accuracy_train.mean())
print(accuracy_test.mean())
```
