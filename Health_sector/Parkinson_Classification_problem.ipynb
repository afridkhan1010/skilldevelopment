```python
# capping method
```

```python
df_outlier=df.copy()
```

```python
df_outlier
```

```python
for i in df_outlier:
    q1=df_outlier[i].quantile(0.25)
    q3=df_outlier[i].quantile(0.75)
    iqr=q3-q1
    lr=q1-(1.5*iqr)
    ur=q3+(1.5*iqr)
    df_outlier[i]=np.where((df_outlier[i]<lr) | (df_outlier[i]>ur),df_outlier[i].median(),df_outlier[i])
```

```python
plt.figure(figsize=(10,20))
df_outlier.boxplot(vert=0)
plt.show()
```

```python
def remove_outliers(col):
    sorted(col)
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    return lb,ub
```

```python
for column in df.columns:
    lb,ub=remove_outliers(df_outlier[column])
    df_outlier[column]=np.where(df_outlier[column]>ub,ub,df_outlier[column])
    df_outlier[column]=np.where(df_outlier[column]<lb,lb,df_outlier[column])
```

```python
plt.figure(figsize=(10,20))
df_outlier.boxplot(vert=0)
plt.show()
```
